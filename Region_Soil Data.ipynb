```python
# Drop empty column
df = df.drop(columns=[" "], errors="ignore")

# Standardize column names
df.columns = df.columns.str.lower().str.strip()
df
```

```python
df = df.rename(columns={
    "region": "state",
    "formatted address": "district",
    "crop": "crop_raw"
})
df
```

```python
import numpy as np

def percent_to_float(x):
    if isinstance(x, str):
        return float(x.replace("%", "").strip())
    return np.nan

for col in df.columns:
    if "%" in str(df[col].iloc[0]):
        df[col] = df[col].apply(percent_to_float)

df
```

```python
df["n"] = (
    df["nitrogen - high"] * 3 +
    df["nitrogen - medium"] * 2 +
    df["nitrogen - low"] * 1
) / 100

df["p"] = (
    df["phosphorous - high"] * 3 +
    df["phosphorous - medium"] * 2 +
    df["phosphorous - low"] * 1
) / 100

df["k"] = (
    df["potassium - high"] * 3 +
    df["potassium - medium"] * 2 +
    df["potassium - low"] * 1
) / 100

df["ph"] = (
    df["ph - acidic"] * 5 +
    df["ph - neutral"] * 7 +
    df["ph - alkaline"] * 9
) / 100

df
```

```python
df["crop"] = df["crop_raw"].str.split(",").str[0].str.lower().str.strip()
df
```
