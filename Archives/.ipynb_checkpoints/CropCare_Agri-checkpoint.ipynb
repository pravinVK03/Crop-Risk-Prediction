```python
# ==========================================
# CLEANING: INDIA CROP PRODUCTION DATASET
# ==========================================

import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("State_District Level Production_Data.csv")
# Rename columns (based on your exact schema)
df = df.rename(columns={
    "State_Name": "state",
    "District_Name": "district",
    "Crop_Year": "year",
    "Season": "season",
    "Crop": "crop",
    "Area": "area",
    "Production": "production"
})

# Convert column names to lowercase (safety)
df.columns = df.columns.str.lower()

# Clean text fields
for col in ["state", "district", "crop", "season"]:
    df[col] = df[col].astype(str).str.strip().str.lower()

# Handle missing values
df = df.dropna(subset=["state", "district", "crop", "year"])

df["area"] = df["area"].fillna(df["area"].median())
df["production"] = df["production"].fillna(df["production"].median())

# Remove invalid entries
df = df[(df["area"] > 0) & (df["production"] > 0)]

# Ensure valid year range
df = df[(df["year"] >= 1997) & (df["year"] <= 2021)]

# Feature engineering: Yield
df["yield"] = df["production"] / df["area"]

# Normalize crop names (important for merging later)
crop_map = {
    "paddy": "rice",
    "rice": "rice",
    "wheat": "wheat",
    "maize": "maize",
    "sugarcane": "sugarcane",
    "cotton(lint)": "cotton",
    "cotton": "cotton"
}
df["crop"] = df["crop"].replace(crop_map)

# Final validation
print("✅ Dataset Cleaned Successfully")
print("Shape:", df.shape)

print("\nSample Data:")
print(df.head())

print("\nUnique Seasons:", df["season"].unique())
print("\nTop Crops:")
print(df["crop"].value_counts().head())
```
