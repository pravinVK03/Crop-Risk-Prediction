```python
#India District-wise Agriculture Crop Production (1997–2021)
```

```python
# STEP 1: LOAD DATA

import pandas as pd

df = pd.read_csv("State_District Level Production Data.csv")

print("Shape:", df.shape)
print("\nColumns:", df.columns)
print("\nSample:")
df.head()
```

```python
# STEP 2: RENAME COLUMNS

df = df.rename(columns={
    "State_Name": "state",
    "District_Name": "district",
    "Crop_Year": "year",
    "Season": "season",
    "Crop": "crop",
    "Area": "area",
    "Production": "production"
})
df
```

```python
# STEP 3: TEXT CLEANING

text_cols = ["state", "district", "crop", "season"]

for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.lower()

# Remove special characters if any
df["state"] = df["state"].str.replace(r"[^a-z\s]", "", regex=True)
df["district"] = df["district"].str.replace(r"[^a-z\s]", "", regex=True)
df
```

```python
# STEP 4: HANDLE MISSING VALUES

# Drop critical missing values
df = df.dropna(subset=["state", "district", "crop", "year"])

# Fill numerical columns
df["area"] = df["area"].fillna(df["area"].median())
df["production"] = df["production"].fillna(df["production"].median())
df
```

```python
# STEP 5: REMOVE INVALID VALUES

# Remove zero/negative values
df = df[(df["area"] > 0) & (df["production"] > 0)]

# Filter valid year range
df = df[(df["year"] >= 1997) & (df["year"] <= 2021)]

# ROBUST OUTLIER CLIPPING
# AREA
area_low = df["area"].quantile(0.02)
area_high = df["area"].quantile(0.98)

df["area"] = df["area"].clip(area_low, area_high)

# PRODUCTION
prod_low = df["production"].quantile(0.02)
prod_high = df["production"].quantile(0.98)

df["production"] = df["production"].clip(prod_low, prod_high)

df
```

```python
# STEP 6: REMOVE DUPLICATES

df = df.drop_duplicates().reset_index(drop=True)
df
```

```python
# EXTRACT UNIQUE CROP NAMES

unique_crops = sorted(df["crop"].dropna().unique())

print("Total Unique Crops:", len(unique_crops))
print("\n--- Crop List ---\n")

for i, crop in enumerate(unique_crops, 1):
    print(f"{i}. {crop}")
```

```python
#STEP 7: NORMALISATION
#Crop normalization

crop_map = {

    # 70, 84 → Rice
    "paddy": "rice",                 # 70
    "rice": "rice",                  # 84

    # 30, 47 → Cotton
    "cotton(lint)": "cotton",        # 30
    "kapas": "cotton",               # 47

    # Pulses
    "moong(green gram)": "moong",    # 57
    "blackgram": "urad",             # 13
    "urad": "urad",                  # 104
    "arhar/tur": "tur",            # 3
    "gram": "gram",                  # 37
    "lentil": "lentil",              # 51
    "masoor": "masoor",              # 55
    "peas & beans (pulses)": "pulses", # 72
    "cowpea(lobia)": "cowpea",       # 31

    # Oilseeds
    "groundnut": "groundnut",        # 39
    "sesamum": "sesamum",            # 91
    "soyabean": "soybean",           # 93
    "sunflower": "sunflower",        # 95
    "rapeseed &mustard": "rapeseed_mustard", # 82
    "niger seed": "niger",           # 59
    "castor seed": "castor",         # 22
    "linseed": "linseed",            # 52

    # Cereals & Millets
    "bajra": "bajra",                # 5
    "jowar": "jowar",                # 44
    "ragi": "ragi",                  # 80
    "maize": "maize",                # 53
    "wheat": "wheat",                # 106
    "barley": "barley",              # 7
    "small millets": "millets",      # 92
    "samai": "millets",              # 88
    "varagu": "millets",             # 105
    "korra": "millets",              # 49

    # Plantation / Commercial
    "sugarcane": "sugarcane",        # 94
    "coffee": "coffee",              # 26
    "tea": "tea",                    # 98
    "tobacco": "tobacco",            # 99
    "rubber": "rubber",              # 86

    # Spices
    "dry chillies": "chillies",      # 33
    "turmeric": "turmeric",          # 102
    "ginger": "ginger",              # 36
    "dry ginger": "ginger",          # 34
    "black pepper": "pepper",        # 12
    "cardamom": "cardamom",          # 17
    "coriander": "coriander",        # 29

    # Fruits
    "banana": "banana",              # 6
    "mango": "mango",                # 54
    "grapes": "grapes",              # 38
    "orange": "orange",              # 62
    "lemon": "lemon",                # 50
    "papaya": "papaya",              # 71
    "pineapple": "pineapple",        # 74
    "sapota": "sapota",              # 90
    "pome granet": "pomegranate",    # 76

    # Vegetables
    "onion": "onion",                # 61
    "potato": "potato",              # 77
    "tomato": "tomato",              # 100
    "cabbage": "cabbage",            # 15
    "cauliflower": "cauliflower",    # 23
    "brinjal": "brinjal",            # 14
    "bhindi": "okra",                # 10
    "pump kin": "pumpkin",           # 79
    "bottle gourd": "bottle gourd",         # 14 (approx group)
    "bitter gourd": "bitter gourd",         # 11

    # Nuts / Plantation
    "arecanut": "arecanut",          # 2
    "arcanut (processed)": "arecanut", # 1
    "atcanut (raw)": "arecanut",     # 4
    "cashewnut": "cashew",           # 19
    "cashewnut raw": "cashew",       # 21
    "cashewnut processed": "cashew", # 20
    "coconut": "coconut",            # 25

    # Others (useful)
    "jute": "jute",                  # 45
    "jute & mesta": "jute",          # 46
    "mesta": "mesta",                # 56
    "safflower": "safflower",        # 87
    "guar seed": "guar",             # 40
    "horse-gram": "horsegram",       # 41

    # ❌ REMOVE (aggregate / noisy categories)
    "total foodgrain": None,         # 101
    "oilseeds total": None,          # 60
    "pulses total": None,            # 78
    "other oilseeds": None,          # 68
    "other vegetables": None,        # 69
    "other fresh fruits": None,      # 65
    "other cereals & millets": None, # 64
    "other kharif pulses": None,     # 66
    "other rabi pulses": None,       # 63
    "other misc. pulses": None,      # 67
    "cond-spcs other": None          # 28
}
df["crop"] = df["crop"].replace(crop_map)

#Season Normalization
season_map = {
    "kharif": "kharif",
    "rabi": "rabi",
    "summer": "summer",
    "whole year": "annual",
    "winter": "rabi",
    "autumn": "kharif"
}

df["season"] = df["season"].map(season_map)

# Drop unwanted categories
df = df[df["crop"].notna()]
df = df[df["season"].notna()]
df
```

```python
# STEP 8: FEATURE ENGINEERING
import numpy as np
# Yield calculation
df["yield"] = df["production"] / df["area"]
lower = df["yield"].quantile(0.01)
upper = df["yield"].quantile(0.99)

df["yield"] = df["yield"].clip(lower, upper)

df["yield_log"] = np.log1p(df["yield"])

# CREATE RISK LABEL
low = df["yield_log"].quantile(0.33)
high = df["yield_log"].quantile(0.66)

def risk_label(y):
    if y < low:
        return 2   # High Risk
    elif y < high:
        return 1   # Medium Risk
    else:
        return 0   # Low Risk

df["risk"] = df["yield_log"].apply(risk_label)
df
```

```python
# STEP 9: VALIDATION

print("Final Shape:", df.shape)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nUnique Seasons:", df["season"].unique())

print("\nTop Crops:")
print(df["crop"].value_counts().head())

print("\nSummary:")
df.describe()
```

```python
# STEP 10: CURATED DATASET
import pandas as pd
final_df = df[[
    "state",
    "district",
    "year",
    "season",
    "crop",
    "area",
    "production",
    "yield_log",
    "risk"
]].reset_index(drop=True)

final_df.to_csv("Dataset_X.csv", index=False)

print("Final curated dataset saved")

df=pd.read_csv("Dataset_X.csv")
df
```
